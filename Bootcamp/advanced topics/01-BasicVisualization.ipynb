```python
# "plt" will be our access point to the main figure and axis
# plots can get much more intricate, but well cover viz in more detail in another week
import matplotlib.pyplot as plt
```

```python
plt.plot([1, 10])
# x-axis is assumed based on the index of the number
# line is interpolated
```

```python
plt.plot([2, 8], [1, 10])  # two separate lists for x and y parts of the coordinates
```

```python
import numpy as np

x = np.linspace(0, 10)  # 0 to 10 (inclusive), broken in to 50 pieces
x
```

```python
y = np.sin(x)
y
```

```python
plt.plot(x, y)
```

```python
plt.plot(x, y, 'mh')  # Magenta hexagons
```

```python
plt.plot(x, y, 'mh')
plt.axis([0, 2 * np.pi, -1, 1])  # x-min, x-max, y-min, y-max
```

```python
plt.plot(x, y, 'mh')
plt.axis([0, 2 * np.pi, -1, 1])
plt.ylabel('Amplitude')
plt.xlabel('Radians')
plt.title('Sine Wave')
```

```python
x = np.random.random_sample(50)
y = np.random.random_sample(50)

plt.scatter(x, y)
```

```python
plt.subplot(1, 2, 1)  # nrows, ncols, plot_number
plt.plot(np.sort(y))

plt.subplot(122)  # abbriviation for x, y, z
plt.hist(y)
```

```python
plt.figure(figsize=(15, 5))  # Adjust size of figure. 

# These are just plots within one figure!
plt.subplot(1, 2, 1)
plt.plot(np.sort(y))

plt.subplot(1, 2, 2)
plt.hist(y)

# There's no some notion of a figure that is separate from the idea of a plot.
```

```python
plt.figure(figsize=(8, 8))

plt.plot(
    np.arange(len(y)), 
    np.linspace(0, 1),

    np.arange(len(x)), 
    np.sort(x), 
    
    np.arange(len(y)), 
    np.sort(y), 
)

plt.legend(('control', 'x', 'y'))
```
